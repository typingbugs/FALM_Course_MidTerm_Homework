```python
from datasets import load_dataset
import os
import json
```

```python
hf_dataset = load_dataset(
    'raw_data/IWSLT/iwslt2017/iwslt2017.py', 
    name='iwslt2017-en-de', 
    trust_remote_code=True
)
```

```python
for split in ["train", "validation", "test"]:
    subset = hf_dataset[split]
    data = [datum['translation'] for datum in subset]
    output_path = os.path.join("iwslt2017-en-de", f"{split}.jsonl")
    print(f"Saving {split} subset to {output_path} ...")
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    with open(output_path, "w", encoding="utf-8") as f:
        for item in data:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")
```
